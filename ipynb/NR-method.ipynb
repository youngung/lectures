$ f(x)=cos(x)+3sin(x)+3\exp(-2x)=0 $ 의 해를 찾아보자.

$ \partial f(x)/\partial x = -\sin(x) + 3\cos(x)-6\exp(-2x)$

```python
def f(x):
    return np.cos(x)+3*np.sin(x)+3*np.exp(-2*x)
def df(x):
    return -np.sin(x)+3*np.cos(x)-6*np.exp(-2*x)
```

```python
x=linspace(0,10)
y=np.cos(x)+3.*np.sin(x)+3.*np.exp(-2.*x)


fig=plt.figure(figsize=(6,5))
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.set_xlabel(r'$x$',fontsize=20)
ax.set_ylabel(r'$y, f(x)$',fontsize=20)
ax.grid()
ax.set_ylim(-10,5)
fig.savefig('init.pdf')
```

# Let's guess x.

```python
guess=7
it=0

x=linspace(0,10)
y=np.cos(x)+3.*np.sin(x)+3.*np.exp(-2.*x)


fig=plt.figure(figsize=(6,5))
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.plot(guess,f(guess),'x')
ax.set_xlabel(r'$x$',fontsize=20)
ax.set_ylabel(r'$y, f(x)$',fontsize=20)
ax.set_ylim(-10,5)
ax.grid()
fig.savefig('%i.pdf'%it)
```

```python
it=it+1
x=linspace(0,10)
y=np.cos(x)+3.*np.sin(x)+3.*np.exp(-2.*x)
xs=np.array([0,10])

## slope line function
g = lambda x: df(guess)*(x-guess)+f(guess)
ys= g(xs)

newx=guess+-f(guess)/df(guess)


fig=plt.figure(figsize=(6,5))
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.plot(guess,f(guess),'x',zorder=-1)
ax.plot(xs,ys,'--')
plot(newx,g(newx),'o')
plot([newx,newx],[0,f(newx)],'--')
plot(newx,f(newx),'x',color='orange')


ax.set_xlabel(r'$x$',fontsize=20)
ax.set_ylabel(r'$y, f(x)$',fontsize=20)
ax.grid()
ax.set_ylim(-10,5)
fig.savefig('%i.pdf'%it)

guess=newx
```
