## example 1

```python
x=np.linspace(-4,4)
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,+2)
ax.set_xlabel('x')
ax.set_ylabel('y')



y=np.exp(-x**2)
ax.plot(x,y,label=r'$e^{-x^2}$')
legend(loc='upper left')
fig.savefig('0.png',dpi=300)

y=x.copy();y[::]=1.
ax.plot(x,y,label=r'$y=1$')
legend(loc='upper left')
fig.savefig('1.png',dpi=300)

y=1.-x**2
ax.plot(x,y,label=r'$y=1-x^2$')
legend(loc='upper left')
fig.savefig('2.png',dpi=300)


#y=1.-x**2+1/2*x**4
#ax.plot(x,y,label=r'$y=1-x^2+\frac{1}{2}x^4$')
#legend()
#fig.savefig('2.png')
```

## example 2

```python
x=np.linspace(-25,25,1000)

fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_xlim(-5,5)
ax.set_ylim(0,+20)
ax.set_xlabel('x')
ax.set_ylabel('y')


y=np.exp(x)
ax.plot(x,y,label=r'$e^{x}$')
legend(loc='upper left')
fig.savefig('0.png',dpi=300)

y=x.copy();y[::]=1.+x
ax.plot(x,y,label=r'$y=1+x$')
legend(loc='upper left')
fig.savefig('1.png',dpi=300)

y=1.+x+0.5*x**2
ax.plot(x,y,label=r'$y=1+x+\frac{1}{2}x^2$')
legend(loc='upper left')
fig.savefig('2.png',dpi=300)

y=1.+x+0.5*x**2+1./6*x**3
ax.plot(x,y,label=r'$y=1+x+\frac{1}{2}x^2+\frac{1}{6}x^3$')
legend(loc='upper left')
fig.savefig('3.png',dpi=300)
```

```python
def gen_lab(a,b,c):
    lab='y(x)='
    if a!=0:
        lab='%s+%s'%(lab,a)
    if b!=0:
        lab='%s+%sx'%(lab,b)
    if c!=0:
        lab='%s+%sx^2'%(lab,c)
        
    if a==0 and b==0 and c==0: lab='y(x)=0'
    return r'$%s$'%lab
```
