```python
def gen_lab(a,b,c):
    lab='y(x)='
    if a!=0:
        lab='%s+%s'%(lab,a)
    if b!=0:
        lab='%s+%sx'%(lab,b)
    if c!=0:
        lab='%s+%sx^2'%(lab,c)
        
    if a==0 and b==0 and c==0: lab='y(x)=0'
    return r'$%s$'%lab
```

```python
def Y(x):
    return np.cos(x+np.pi/3.)
def dY(x):
    return -np.sin(x+np.pi/3.)
def d2Y(x):
    return -np.cos(x+np.pi/3.)
def d3Y(x):
    return np.sin(x+np.pi/3.)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
x=np.linspace(-np.pi*3,np.pi*1.5,100)
ax.set_ylim(-1.5,2.)

y=Y(x)
ax.plot(x,y,label=r'$y(x)=cos(x+\frac{\pi}{3})$')
ax.legend(loc='upper left',fontsize=12)
fig.savefig('/tmp/0.png',dpi=300)

X=np.zeros(100)
y=np.linspace(-1.5,2,X.shape[0])
l,=ax.plot(X,y,'--',c='gray',lw=1)
fig.savefig('/tmp/1.png',dpi=300)

a=Y(0.)
b=0
c=0
lab=gen_lab(a,b,c)
p=np.poly1d([c,b,a])
y=p(x)
ax.plot(x,y,label=r'$P(x)=y(0)$')  ##cos(%3.1f+\frac{\pi}{3})$'%a)
ax.legend(loc='upper left',fontsize=12)
fig.savefig('/tmp/2.png',dpi=300)


a=Y(0.)
b=dY(0.)
c=0
lab=gen_lab(a,b,c)
p=np.poly1d([c,b,a])
y=p(x)
ax.plot(x,y,label=r'$P(x)=y(0)+y^\prime(0)x$')  ##cos(%3.1f+\frac{\pi}{3})$'%a)
ax.legend(loc='upper left',fontsize=12)
fig.savefig('/tmp/3.png',dpi=300)



a=Y(0.)
b=dY(0.)
c=0.5*d2Y(0.)
lab=gen_lab(a,b,c)
p=np.poly1d([c,b,a])
y=p(x)
#ax.plot(x,y,label=r'$P(x)=y(0)+y^\prime(0)x%+3.2fx^2$'%c)
ax.plot(x,y,label=r'$P(x)=y(0)+y^\prime(0)x+\frac{1}{2}y^{\prime\prime}(0)x^2$'%c) 
ax.legend(loc='lower left',fontsize=12)
fig.savefig('/tmp/4.png',dpi=300)


a=Y(0.)
b=dY(0.)
c=0.5*d2Y(0.)
d=1/6.*d3Y(0.)
#lab=gen_lab(a,b,c,d)
p=np.poly1d([d,c,b,a])
y=p(x)
ax.plot(x,y,label=r'$P(x)=y(0)+y^\prime(0)x+\frac{1}{2}y^{\prime\prime}(0)x^2+\frac{1}{6}y^{\prime\prime\prime}(0)x^3$'%c) 
ax.legend(loc='lower left',fontsize=12)
fig.savefig('/tmp/5.png',dpi=300)


"""

a=Y(0.)
b=dY(0.)
c=0
lab=gen_lab(a,b,c)
p=np.poly1d([c,b,a])
y=p(x)
ax.plot(x,y,label=r'$P(x)=P(0)%+5.3fx$'%b)  ##cos(%3.1f+\frac{\pi}{3})$'%a)
ax.legend(loc='upper left',fontsize=12)
fig.savefig('/tmp/3.png',dpi=300)"""
```
