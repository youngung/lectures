```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(12,3))
ax1 = fig.add_subplot((121),projection='3d')
ax2 = fig.add_subplot(122)

# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(0, 1., 50)
p = np.linspace(0, 2*np.pi, 50)
R, P = np.meshgrid(r, p)
Z = ((R-1.)**2)

x=np.linspace(-1,1,1000)

z=((abs(x)-1.)**2)

# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)

# Plot the surface.
ax1.plot_surface(X, Y, Z, cmap=plt.cm.YlGnBu_r)
ax2.plot(x,z,'-k')
ax2.set_xlabel('x')# [mm]')
ax2.set_ylabel('$v_z$')# [mm/s]')

# Tweak the limits and add latex math labels.
#ax.set_zlim(0, 1)
#ax.set_xlabel(r'$\phi_\mathrm{real}$')
#ax.set_ylabel(r'$\phi_\mathrm{im}$')
ax1.set_zlabel(r'$V_z$')# [mm/s]')
ax1.set_axis_off()
ax2.set_xticks(np.linspace(-1,1,5))

xs=np.linspace(-1,1,15)
for i in range(len(xs)):
    x=xs[i]
    y=(abs(x)-1.)**2
    #ax2.plot([x,x],[0,y],'-k')
    ax2.arrow(x,0,0,y,shape='full',width=0.01,length_includes_head=True,color='r')
fig.tight_layout()
fig.savefig('/tmp/prof.pdf')
```
